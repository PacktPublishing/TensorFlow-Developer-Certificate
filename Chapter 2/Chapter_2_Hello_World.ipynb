```python
# import tensorflow 
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
print(tf.__version__)
```

```python
#import additional libraries
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Hours of study
X = [20,23,25,28,30,37,40,43,46]
  
# Test Scores
y = [40,51,55,61,65,70,85,91,97]
  
```

```python
plt.plot(X, y)
plt.title("Exam Performance graph")
plt.xlabel('Hours of Study')
plt.ylabel('Test Score')
plt.show()
```

```python
study_model = Sequential([Dense(units=1,input_shape=[1])])
study_model.compile(optimizer='adam', loss='mean_squared_error')

X= np.array(X, dtype=int)
y= np.array(y, dtype=int)
```

```python
#fitting the model 
history= study_model.fit(X,y, epochs=2500)
```

```python
print(history.history.keys())
# Lets plot the loss
plt.plot(history.history['loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Number of epochs')
plt.legend(['loss plot'], loc='upper right')
plt.show()
```

```python
#Let us predict how well a student will performance based on their study time

n=38  #Hours of study
result =study_model.predict([n])[0][0] #Result
result =round(result) #Rounding off the results as a whole number
```

```python
print(f"If I study for {n} hours, I will get {result} marks as my grade.")
```
